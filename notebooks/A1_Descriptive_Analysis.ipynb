<p><b>Задача 2.</b> Анализ рынка акций (4 балла)</p>

<p>1. Загрузите данные в один dataframe из всех файлов в папке /data/stock. Все файлы имеют одинаковую структуру, в том числе наименование столбцов. В качестве значений индекса dataframe'а необходимо указать значения столбца "Date". Название столбцов должные соответствовать названию акций (имя файла без .csv), а их значения - значениям цены закрытия (столбец "Close" в файлах .csv)</p>

```python
# Ваш код
import os
path = "../data/A1_Descriptive_Analysis/stock"   # Путь к файлам
full_file_name = os.listdir(path) #список файлов в папке file_path
Matr = pd.DataFrame() #пустой датафрейм
for filename in full_file_name: # для всех имен
    buffer = pd.read_csv(path+"/"+filename) #считываем в буфер файл
    Matr[filename.partition('.')[0]] = buffer["Close"] #для нового столбца с названием как у файла но без расширения записываем столбец буффера закрытия
    indexM = buffer["Date"] # идекс для матрицы дата
Matr.index = indexM
Matr
```

```python
# Ваш код
Matr.corr()
```

```python
# Ваш код
f = plt.figure(figsize=(10, 7))
plt.matshow(Matr.corr(), fignum=f.number, vmin=-1, vmax=1) # рисуем диаграмму
plt.xticks(range(Matr.shape[1]), Matr.columns, fontsize=8, rotation=90) # подписи по осям
plt.yticks(range(Matr.shape[1]), Matr.columns, fontsize=8)
cb = plt.colorbar() # шкала справа
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix Diagram')
plt.show()
```

<p><b>Вариант 1.</b> Microsoft (MSFT)</p>
<p><b>Вариант 2.</b> Facebook (FB)</p>
<p><b>Вариант 3.</b> Google (GOOGL)</p>
<p><b>Вариант 4.</b> Alibaba (BABA)</p>

```python
# Ваш код
action = Matr.corr()["FB"].drop(index = "FB") # берем значения корреляции для facebook и удаляем корреляцию на саму себя
min = 0
minZnach = 0
for j in action.to_numpy(): # считаем корреляцию близкую к отсутствию
    if (1 - abs(j) > min):
        min = 1 - abs(j)
        minZnach = j
print("Максимальная корреляция :",action.idxmax(),'=',action.loc[action.idxmax()],'\n',
      "Минимальная корреляция :",action.idxmin(),'=',action.loc[action.idxmin()],'\n',
      "Близка к отсутствию корреляции :",action[Matr.corr()["FB"] == minZnach].index.values.tolist()[0],'=',minZnach)
```

```python
# Ваш код
# scatter
plt.figure(figsize=[15, 5])

plt.subplot(1,3,1)
plt.scatter(x = Matr["FB"],y=Matr[action.idxmax()], color = "green")
plt.xlabel("FB")
plt.ylabel(action.idxmax())

plt.subplot(1,3,2)
plt.scatter(x = Matr["FB"],y=Matr[action.idxmin()], color = "orange")
plt.xlabel("FB")
plt.ylabel(action.idxmin())

plt.subplot(1,3,3)
plt.scatter(x = Matr["FB"],y=Matr[action[Matr.corr()["FB"] == minZnach].index.values.tolist()[0]], color = "black")
plt.xlabel("FB")
plt.ylabel(action[Matr.corr()["FB"] == minZnach].index.values.tolist()[0])
#

plt.show()
```
